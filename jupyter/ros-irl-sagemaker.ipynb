```python
import csv
import glob
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline 
```

# Modules

```python
def createDict(color, level):
    df = {}
    idx = 0

    dirpath = glob.glob("../data/" + color + "/" + level + "/*.csv")
    for path in dirpath:
        df[idx] = pd.read_csv(path, header=0, delimiter=",", low_memory=False)
        idx += 1
    return df
```

```python
def addPlot(dic):
    plt.figure(figsize=(20,10))
    plt.axhline(y=10)
    for k, v in dic.items():
        plt.plot(v.clock, abs(v.scores_b - v.scores_r), marker='o', linestyle="solid")
```

# Level1

```python
dic1 = createDict("red", "level1")
addPlot(dic1)
plt.show()
```

# Level2

```python
dic2 = createDict("red", "level2")
addPlot(dic2)
plt.show()
```

# Level 3

```python
dic3 = createDict("red", "level3")
addPlot(dic3)
plt.show()
```
